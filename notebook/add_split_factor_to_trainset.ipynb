```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import plotly.express as px
import warnings;warnings.simplefilter('ignore')
root = '../data/phase_2/'
os.listdir(root)
```

```python
def get_group_col(df,name):
    condition = df.columns.str.contains(name)
    return df.columns[condition].tolist()
```

# Load data

```python
train = pd.read_csv(root+'train.csv',index_col=0).dropna()
train.columns = [i.replace('  ','') for i in  train.columns]
print(train.shape)
train.head()
```

# define columns

```python
case_col = set(get_group_col(train,'Case'))&set(get_group_col(train,'Input'))
case_col = sorted(list(case_col))
case_col
```

# X_na and F_na

```python
X_na = get_group_col(train,'Naphtha Properties')[5:-1]
F_na = get_group_col(train,'Naphtha Properties')[1]
print(len(X_na))
print(X_na[0],X_na[-1])
print(F_na)
```

# X_le,F_le

```python
X_le = get_group_col(train,'Light End Product Properties')[3:-1]
F_le = get_group_col(train,'Light End Product Properties')[1]
print(F_le)
print(len(X_le),X_le[0],'~',X_le[-1])
```

# X_hc,F_hc

```python
X_hc = get_group_col(train,'Heart Cut Product Properties')[4:-1]
F_hc = get_group_col(train,'Heart Cut Product Properties')[1]
print(F_hc)
print(len(X_hc),X_hc[0],'~',X_hc[-1])
```

# X_he,F_he

```python
X_he = get_group_col(train,'Heavy End Product Properties')[3:-1]
F_he = get_group_col(train,'Heavy End Product Properties')[1]
print(F_he)
print(len(X_he),X_he[0],'~',X_he[-1])
```

# caculate_split_factor

```python
a = train[X_le].values * train[F_le].values.reshape(-1,1)
b = train[X_na].values * train[F_na].values.reshape(-1,1)
sle = a / (b+1e-12)
sle = pd.DataFrame(sle,columns=[ 'Light End Split Factor_'+i.split('_')[1] for i in X_na])
sle.head(3)
```

```python
a = train[X_hc].values * train[F_hc].values.reshape(-1,1)
b = train[X_na].values * train[F_na].values.reshape(-1,1)
shc = a / (b+1e-12)
shc = pd.DataFrame(shc,columns=[ 'Heart Cut Split Factor_'+i.split('_')[1] for i in X_na])
shc.head(3)
```

```python
a = train[X_he].values * train[F_he].values.reshape(-1,1)
b = train[X_na].values * train[F_na].values.reshape(-1,1)
she = a / (b+1e-12)
she = pd.DataFrame(she,columns=[ 'Heavy End Split Factor_'+i.split('_')[1] for i in X_na])
she.head(3)
```

```python
train = pd.concat([train,sle,shc,she],axis=1)
train.shape
```

# check split factor

```python
for e,i in enumerate([i.split('_')[1] for i in X_na]):
    col_1 = 'Light End Split Factor_{}'.format(i)
    col_2 = 'Heart Cut Split Factor_{}'.format(i)
    col_3 = 'Heavy End Split Factor_{}'.format(i)
    cols = [col_1,col_2,col_3]
    temp = train[cols]
    temp['SUM'] = temp.sum(axis=1)
    print(e,round(temp['SUM'].mean(axis=0),1),i)
```

```python
for i in ['Oxygen','Methane','Ethane','H2O']:
    col = 'Light End Split Factor_{}'.format(i)
    train[col] = 1.0
```

```python
for i in ['n-Tridecane']:
    col = 'Heavy End Split Factor_{}'.format(i)
    train[col] = 1.0
```

```python
for e,i in enumerate([i.split('_')[1] for i in X_na]):
    col_1 = 'Light End Split Factor_{}'.format(i)
    col_2 = 'Heart Cut Split Factor_{}'.format(i)
    col_3 = 'Heavy End Split Factor_{}'.format(i)
    cols = [col_1,col_2,col_3]
    temp = train[cols]
    temp['SUM'] = temp.sum(axis=1)
    print(e,round(temp['SUM'].mean(axis=0),1),i)
```

```python
train = train.dropna(axis=0)
train.shape
```

```python
train.to_csv(root+'train(add_split_factor).csv')
print('done')
```
